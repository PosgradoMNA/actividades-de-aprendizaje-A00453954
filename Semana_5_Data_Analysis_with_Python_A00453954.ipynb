```python
import pandas as pd
import numpy as np
```

```python
path='https://raw.githubusercontent.com/luisfernandorios/Actividades_Aprendizaje-/main/imports-85.data'
```

```python
df= pd.read_csv(path,sep=',',header=None)
```

```python
df.head(22)
```

```python
headers=['symboling','normalized-losses','make','fuel-type','aspiration','num-of-doors','body-style','drive-wheels','engine-location','wheel-base','length','width','height','curb-weight','engine-type','num-of-cylinders','engine-size','fuel-system','bore','stroke','compression-ratio','horsepower','peak-rpm','city-mpg','highway-mpg','price']
df.columns=headers
```

```python
df=df.replace('?',np.NaN)
```

```python
df.head(50)
```

```python
df=df.dropna(subset=['price'],axis=0)
df.info()
```

```python
df['price']=df['price'].astype('int64')
```

```python
df.dtypes
```

```python
df['symboling']=df['symboling']+1
```

```python
df.head()
```

```python
df['city-mpg']=235/df['city-mpg']
```

```python
df.head()
```

```python
df= df.rename(columns={'city-mpg':'city-L/100km'})
```

```python
df.head()
```

```python
df['normalized-losses'] = df['normalized-losses'].astype('float')
mean_val = df['normalized-losses'].mean()
```

```python
df['normalized-losses']=df['normalized-losses'].replace(np.NaN,mean_val)
df.head(2)
```

```python
##NORMALIZATION
```

```python
#Scaling
df['wheel-base']=df['wheel-base']/df['wheel-base'].max()
```

```python
df['wheel-base'].head(5)
```

```python
#Min-Max
df['horsepower'] = df['horsepower'].astype('float')
df['horsepower']=(df['horsepower']-df['horsepower'].min())/(df['horsepower'].max()-df['horsepower'].min())
df['horsepower'].head(5)
```

```python
#Z-Score
df['peak-rpm'] = df['peak-rpm'].astype('float')
df['peak-rpm']=(df['peak-rpm']-df['peak-rpm'].mean())/df['peak-rpm'].std()
df['peak-rpm'].head(5)
```

```python
#Binning
bins=np.linspace((df['price'].min()),(df['price'].max()),4)
bins
```

```python
group_names=['LOW','MED','HIGH']
```

```python
df['price-bins']=pd.cut(df['price'],bins,labels=group_names,include_lowest=True)
```

```python
df.head(2)
```

```python
import seaborn as sns
import matplotlib as plt
```

```python
sns.histplot(df['price-bins'])
```

```python
df['fuel-type'] = df['fuel-type'].astype('str')
```

```python
sns.histplot(df['fuel-type'])
```

```python
dummy_frame=pd.get_dummies(df['fuel-type'])
```

```python
dummy_frame.head()
```

```python
df_dumm = pd.concat([df,dummy_frame],axis=1)
df_dumm.head()
```
